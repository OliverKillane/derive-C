## Benchmark Parameter Analysis
For deciding on ranges, distributions of values to try.

```python
import plotly.express as px
import polars as pl
```

### Exponential ranges
Want a large number of scale factors.

```python
bases = [2, 3, 5, 7]
multipliers = [2, 3, 5]

pairs = []
values = []
max = 65536
for base in bases:
    for multiplier in multipliers:
        pairs.append((base, multiplier))

        val = base
        while val < max:
            values.append(val)
            val *= multiplier

len(values)
```

```python
unique_values = sorted(set(values))
df = pl.DataFrame({"value": unique_values, "index": range(len(unique_values))})

fig = px.scatter(
    df,
    x="value",
    y="index",
    log_x=True,
    title="Exponential Range Values (Logarithmic X-Axis)",
    labels={"value": "Value (log scale)", "index": "Index"},
    height=600,
)
fig.update_traces(marker=dict(size=4))
fig.show()
```
